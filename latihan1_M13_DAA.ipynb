```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque

graph = {
    'Rektor': ['Warek 1', 'Warek 2'],
    'Warek 1': ['Kaprodi 1'],
    'Warek 2': ['Kaprodi 2', 'Kaprodi 3'],
    'Kaprodi 1': ['Dosen A', 'Dosen B'],
    'Kaprodi 2': ['Dosen D', 'Dosen E'],
    'Kaprodi 3': ['Dosen F', 'Dosen G'],
    'Dosen A': ['Dosen C'],
    'Dosen B': [],
    'Dosen C': [],
    'Dosen D': [],
    'Dosen E': [],
    'Dosen F': [],
    'Dosen G': []
}

def draw_graph(graph):
    G = nx.DiGraph(graph)

    pos = {
        'Rektor': (0, 4),
        'Warek 1': (-2, 3), 'Warek 2': (2, 3),
        'Kaprodi 1': (-3, 2), 'Kaprodi 2': (0, 2), 'Kaprodi 3': (3, 2),
        'Dosen A': (-4, 1), 'Dosen B': (-2, 1),
        'Dosen D': (-1, 1), 'Dosen E': (1, 1),
        'Dosen F': (2, 1), 'Dosen G': (4, 1),
        'Dosen C': (-4, 0)
    }


    plt.figure(figsize=(12, 8))


    nx.draw_networkx_nodes(G, pos, node_size=2000, node_color='lightblue', edgecolors='black')


    nx.draw_networkx_edges(G, pos, arrowsize=20, edge_color='gray')


    nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

    plt.title("Visualisasi Skema Organisasi", fontsize=15)
    plt.axis('off')
    plt.show()

def bfs(graph, start_node):
    """
    Performs Breadth-First Search starting from the given node.
    """

    visited = set()

    queue = deque([start_node])

    bfs_order = []

    while queue:

        current_node = queue.popleft()


        if current_node not in visited:

            visited.add(current_node)
            bfs_order.append(current_node)


            for neighbor in graph.get(current_node, []):
                if neighbor not in visited:
                    queue.append(neighbor)

    return bfs_order


print("## ðŸ“Š Visualisasi Skema Organisasi")
draw_graph(graph)
print("\n" + "-"*50 + "\n")


start_node = 'Rektor'
bfs_result = bfs(graph, start_node)

print(f"## ðŸš€ Hasil Breadth-First Search (BFS) dari '{start_node}'")
print("\nUrutan Kunjungan (Level by Level):")
print(bfs_result)


print("\n--- Analisis Level ---")
print("Level 0 (Root):", ['Rektor'])
print("Level 1:", ['Warek 1', 'Warek 2'])
print("Level 2:", ['Kaprodi 1', 'Kaprodi 2', 'Kaprodi 3'])
print("Level 3:", ['Dosen A', 'Dosen B', 'Dosen D', 'Dosen E', 'Dosen F', 'Dosen G'])
print("Level 4:", ['Dosen C'])
```
